## 1. INTRODUCTION

In the medical field, cancer is an inevitable topic. It is one of the primary causes of morbidity and mortality in the world. Cervical cancer, arising from the cervix — the lower part of the uterus that connects to the vagina. This cancer can affect the deeper tissues of a woman’s cervix and may spread to other parts of the body, often the lungs, liver, bladder, vagina and rectum. As the fourth most common cancer and the fourth most common cause of death from cancer, cervical cancer is one of the most dangerous diseases among women. In 2018, an estimated 570, 000 women were diagnosed with cervical cancer worldwide and about 311,000 women died from the disease. It is important for people to attach great importance to cervical cancer. According to the CDC(Centers for Disease Control and Prevention), there are several risk factors of cervical cancer: Age, High number of sexual partners, early first sexual intercourse, high number of pregnancies, smoking and so on. In our project, we want to find and compare the effect of the recognized risk factors of cervical cancer.


The problem with many cancers is the lack of specificity of symptoms that if noticed, would lead to early diagnosis. Cervical cancer grows slowly, so there’s usually sufficient time to find and treat it before it causes serious problems. Identifying and analyzing risk factors associated with cervical cancer would allow people to be diagnosed sooner if they have those certain risk factors and therefore be treated. When diagnosed, cervical cancer is one of the most successfully treatable forms of cancer, as long as it is detected early and managed effectively. Cancers diagnosed in late stages can also be controlled with appropriate treatment and palliative care. We want to learn about the risk factors associated with cervical cancer to determine the likelihood that someone has or will get cancer.

### 2.3. Data Analysis

Our predictive model determines whether someone has cervical cancer using a variety of feature variables. There are four outcome variables that represent four different cervical cancer diagnostic tests. This is a supervised learning task because there are already labels associated with the data. Since we are determining whether or not someone has cancer using four different outcome variables, this is a binary classification problem with multiple labels. We will average the accuracies and F1 scores from each outcome variable to analyze the effectiveness of the machine learning algorithms we use.

The following are the machine learning algorithms we plan on using:

*   K-Nearest Neighbors: This algorithm works by looking at k training samples closest to the given simple. We are using it because it is relatively fast, simple, easy to adjust and commonly used in classification tasks.
*   Support Vector Machine: This algorithm uses lines to divide classifications of data and maximizes the margin between the line and nearest data points. We are using it because it works well with sparse data, which our dataset has a lot of.
*   Logistic Classification: This algorithm works by limiting the range of the linear function to [0,1] and uses the conditional probability of the input value belonging to the positive class. We are using logistic regression because our target variables are binary, and this algorithm works well for binary classification.
*   Decision Tree Regressor: This algorithm works through a series of if/else questions until a result is determined. We are using this algorithm because our data has a mix of continuous and binary features and it works well with mixed data.

```python
#the risk factors are the features we are going to analyze
features = (data.iloc[0:858, 0:30])

#the four diagnoses are the target variables
hinselmann = data["Hinselmann"]
schiller = data["Schiller"]
citology = data["Citology"]
biopsy = data["Biopsy"]
```

```python
#define a dictionary of estimators for model
estimators = {
    'k-Nearest Neighbor': KNeighborsClassifier(), 
    'Support Vector Machine': LinearSVC(max_iter=1000000),
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier()}
```

```python
#function that splits the data using percentage split (75% train 25% test)
def split_data(features, target):
    X_train, X_test, y_train, y_test = train_test_split(features, target, random_state = 3000)
    return X_train, X_test, y_train, y_test
```

```python
#split the data using percentage split
X_train_h, X_test_h, y_train_h, y_test_h = split_data(features, hinselmann)
X_train_s, X_test_s, y_train_s, y_test_s = split_data(features, schiller)
X_train_c, X_test_c, y_train_c, y_test_c = split_data(features, citology)
X_train_b, X_test_b, y_train_b, y_test_b = split_data(features, biopsy)
```

```python
# preprocess using MinMaxScaler
def preprocessor(train, test):
    scaler = MinMaxScaler().fit(train)
    
    #scale testing and training sets
    X_train_scaled = scaler.transform(train)
    X_test_scaled = scaler.transform(test)

    return X_train_scaled, X_test_scaled
```

```python
#preprocess the data
X_train_h_scaled, X_test_h_scaled = preprocessor(X_train_h, X_test_h)
X_train_s_scaled, X_test_s_scaled = preprocessor(X_train_s, X_test_s)
X_train_c_scaled, X_test_c_scaled = preprocessor(X_train_c, X_test_c)
X_train_b_scaled, X_test_b_scaled = preprocessor(X_train_b, X_test_b)
```

```python
# feature selection begins here

# Univariate Selection
# Using this strategy, we check to see whether there is a statistically significant relationship between each feature and target
# The features that are related with the highest confidence are selected.
# Each feature is individually considered.
# This selection performs an ANOVA.

# Model-Based Feature Selection
# Uses a supervised ML algorithm to judge the importance of each feature and keeps only the most important ones
# The model used for selection doesn't need to be the same model that is used for training
# The feature selection model needs to provide some measure of importance for each feature
# so that they can be ranked by this measure

# Iterative Feature Selection
# A series of models are built, with varying numbers of features

```

```python
#define dictionary of feature selection methods
feat_select_dict = {"UNI" : SelectKBest(score_func=f_classif, k = 3),
                    "MB" : SelectFromModel(DecisionTreeRegressor(random_state = 3000)),
                   "RFE" : RFE(DecisionTreeRegressor(random_state = 3000), n_features_to_select = 3)}
```

```python
#set up dataframes to hold results of selecting features vs using all of the features
prelim_train_results = pd.DataFrame(index = ["Acc_All", "Acc_UNI", "Acc_MB", "Acc_RFE"])
prelim_test_results = pd.DataFrame(index = ["Acc_All", "Acc_UNI", "Acc_MB", "Acc_RFE"])
```

```python
# Feature selection for each target variable and each selection method
def feature_selection(feat_select_dict, xtrain, xtest, ytrain, ytest, target):
    ml_model_all_train = LogisticRegression().fit(X=xtrain, y=ytrain)
    ml_model_all_test = LogisticRegression().fit(X=xtest, y=ytest)
    prelim_train_results.loc["Acc_All", target] = ml_model_all_train.score(xtrain, ytrain)
    prelim_test_results.loc["Acc_All", target] = ml_model_all_test.score(xtest, ytest)
    models = []
    
    for name, method in feat_select_dict.items():
        model = method
        model.fit(xtrain, ytrain)
        xtrain_selected = model.transform(xtrain)
        xtest_selected = model.transform(xtest)
        models.append(model)
        ml_model_selected_train = LogisticRegression().fit(X=xtrain_selected, y=ytrain)
        ml_model_selected_test = LogisticRegression().fit(X=xtest_selected, y=ytest)
        prelim_train_results.loc["Acc_" + name, target] = ml_model_selected_train.score(xtrain_selected, ytrain)
        prelim_test_results.loc["Acc_" + name, target] = ml_model_selected_test.score(xtest_selected, ytest)
    return prelim_train_results, prelim_test_results, models
```

```python
prelim_train_results, prelim_test_results, models_h = feature_selection(feat_select_dict, X_train_h_scaled, X_test_h_scaled, y_train_h, y_test_h, "Hinselmann")
prelim_train_results, prelim_test_results, models_s = feature_selection(feat_select_dict, X_train_s_scaled, X_test_s_scaled, y_train_s, y_test_s, "Shiller")
prelim_train_results, prelim_test_results, models_c = feature_selection(feat_select_dict, X_train_c_scaled, X_test_c_scaled, y_train_c, y_test_c, "Citology")
prelim_train_results, prelim_test_results, models_b =  feature_selection(feat_select_dict, X_train_b_scaled, X_test_b_scaled, y_train_b, y_test_b, "Biopsy")
```

```python
hypothesis_data = data[["Age", "Number of sexual partners", "First sexual intercourse", "Num of pregnancies", "Smokes (years)", "Hormonal Contraceptives (years)", "Hinselmann", "Schiller", "Citology", "Biopsy"]]
```

```python
# since we are doing T Test, we have to make sure that all the columns in our dataset are binary
# to transfer the columns into binary values
import numpy as np 

age_range_0 = np.arange(12, 40)
age_range_1 = np.arange(39, 85)

sex_part_0 = np.arange(0, 5)
sex_part_1 = np.arange(5, 29)

first_sex_0 = np.arange(16)
first_sex_1 = np.arange(15, 33)

num_preg_0 = np.arange(4)
num_preg_1 = np.arange(3, 12)

#CLEANING: clean up strings so they can used as numbers
for columns in selected_data:
    selected_data[columns] = pd.to_numeric(selected_data[columns])
    
def trans_to_binary(data):
    smoke_range = data["Smokes (years)"]

    for years in smoke_range:
        if years < 1:
            data["Smokes (years)"].replace(years, 0, inplace=True)
        else:
            data["Smokes (years)"].replace(years, 1, inplace=True)
        
    hor_range = data['Hormonal Contraceptives (years)']

    for years in hor_range:
        if years < 1:
            data['Hormonal Contraceptives (years)'].replace(years, 0, inplace=True)
        else:
            data['Hormonal Contraceptives (years)'].replace(years, 1, inplace=True)
    data["Age"].replace(age_range_0, 0, inplace=True)
    data["Age"].replace(age_range_1, 1, inplace=True)
    data["Number of sexual partners"].replace(sex_part_0, 0, inplace=True)
    data["Number of sexual partners"].replace(sex_part_1, 1, inplace=True)
    data["First sexual intercourse"].replace(first_sex_0, 0, inplace=True)
    data["First sexual intercourse"].replace(first_sex_1, 1, inplace=True)
    data["Num of pregnancies"].replace(num_preg_0, 0, inplace=True)
    data["Num of pregnancies"].replace(num_preg_1, 1, inplace=True)
    return data

hp_test = trans_to_binary(hypothesis_data)
hp_test
```

```python
def f_test(data,IV,DV):
    catego = [x[1] for x in data.groupby(IV)[DV]]
    f_statistics, p_values = stats.f_oneway( * catego)
    df1 = len(data.groupby(IV)[DV]) - 1
    df2 = sum([len(x[1])-1 for x in data.groupby(IV)[DV]])
    print(str(IV) + " ---- F-test")
    print("------------------------------" + "\n")
    print("F("+format(df1,'d')+","+format(df2,'d')+") = " + format(f_statistics,'.2f') + " , p = "+format(p_values,'.4f')+"\n")
    print("------------------------------" + "\n")

def t_test(data,IV,DV):
    results = stats.ttest_rel(data[IV],data[DV])
    #t value
    tstatistic = results[0]
    #p value in scientific notation
    pvalue = results[1]
    print(str(IV) + " ---- t-test")
    print("------------------------------" + "\n")
    print("t-value ="+format(tstatistic,'.2f')+"   pvalue = "+format(pvalue,'.10f')+"\n")
    if(pvalue<0.05):
        print("pvalue is not significant\n")
        print("------------------------------" + "\n")

    else:
        print("pvalue is significant\n")
        print("------------------------------" + "\n")

def check_levene(data,IV,DV):
    catego = [x[1] for x in data.groupby(IV)[DV]]
    levene_results = stats.levene( * catego)
    print(str(IV) +  "\n" + " Assumption of Equality of Variances:")
    print("\t", end = "")
    print(levene_results)
    if (levene_results[1] > 0.05):
        print("\t\t"+"Assumption is met. p > .05" + "\n")
    else:
        print("\t\t"+"Assumption is not met, p < 0.05" + "\n")

        
def check_shapiro(data,IV,DV):
    shapiro = data.groupby(IV)[DV].agg(lambda x : stats.shapiro(x))
    print(str(IV) +  "\n" + "Assumption of Normality:")
    for i in range (shapiro.size):
        print("\n",end = "")
        print(shapiro.index[i], end = " : ")
        print(end = "")
        print(shapiro.iloc[i])
        if(shapiro.iloc[i][1]>0.05):
            print("\t\t"+"Assumption is met. p > .05"+  "\n"+  "\n")
        else:
            print("\t\t"+"Assumption is not met, p < 0.05"+  "\n"+  "\n")
```

## 4. DISCUSSION

To analyze our data, we first removed columns from the dataset that had unknown values in at least half of its rows and changed remaining unknown values to 0. Then we split our data into training and testing sets and scaled the data using MinMaxScaler. Using the scaled data, we applied three different feature selection methods for each of the four target variables and extracted the most important features found in all cases. We also analyzed our data using barplots to visualize the effects of each feature variable on each target variable as well as a heatmap to visualize the correlations between all the variables. To test our predictive model model, the scaled data and the scaled data with the selected features were used in four different machine learning algorithms: KNearest Neighbors, Support Vector Machines, Logistic Regression, and Decision Tree Classifier. The effectiveness of each algorithm was determined using accuracy, F1 score, and prediction of true positives and negatives. The most effective algorithms were then hypertuned using Grid Search.


We compared the KNearest Neighbors, Support Vector Machines, Logistic Regression, and Decision Tree Classifier algorithms. The Logistic Regression algorithm had the best overall performance with the highest accuracy rate and F1 scores when the selected features were used, while KNearest Neighnors was the highest performing when all the features were used. Support Vector Machines ranked as the second best algorithm in both of these instances. We chose the Logistic Regression algorithm for our predictive model because it had the best performance and works well when predicting binary values. Based on our findings, the features from our dataset can accurately predict the four outcome variables (if 0) when using the four classification algorithms mentioned previously. Only Decision Tree could predict true positives, and even then, very few were predicted.


However, when examining the relation between our selected feature variables and the four outcome variables using the ttest, a majority came out with non-significant differences meaning that we could not reject our null hypotheses. If we had significant results, we could reject the null hypothesis and establish that certain risk factors of cervical cancer have an effect on whether someone has cervical cancer. In all cases, at least one of the assumptions (equality of variances and normality) was also not met so the results of the hypothesis tests are not valid. Given these results we remain uncertain about the correlation between the given cervical cancer risk factors in the dataset and the results of cervical cancer diagnostic tests.


Through the analysis of the cervical cancer dataset and testing of our predictive model, we believe it would be beneficial to get more data with less unknown values to more accurately determine the relationship between the feature and target variables. Further hypothesis testing is needed since the assumptions for the t-test were not met. It would probably be best to use a nonparametric test such as the Mann-Whitney U test. The dataset we used contains data only from Venezuela, so it would also be helpful to get data from other countries for a more comprehensive dataset. This dataset also needs to be balanced as there are far more 0s than 1s in the target variable columns.
